## Medidas de Tendência Central

#### Média Aritmética

Abaixo está um teste da função `arithmetic_mean`. A função recebe uma lista de valores 'data' e retorna a média aritmética desses valores.

```python
# Em uma amostra foram registradas a duração (em minutos) de 10 chamadas VoIP: 
data = [2.5, 3.5, 2.8, 1.9, 2.4, 2.6, 3.4, 2.8, 3.6, 2.5]
    
mean = arithmetic_mean(data)

print(f'Duração média de uma chamada VoIP: {mean} min')
plot_values_with_mean(data, mean)
```

**O que a média aritmética ajuda a identificar/revela?**

A média aritmética é uma das medidas de tendência central mais simples e amplamente utilizadas. Ela ajuda a identificar o valor central de um conjunto de dados, revelando o valor típico ou "médio" das observações. 

**A média aritmética é indicada para quais tipos de amostra?**

- **Distribuição Simétrica:** Quando os dados não são significativamente assimétricos e não possuem outliers extremos que possam distorcer o valor da média.
- **Dados Numéricos:** Quando os dados são quantitativos e contínuos.
- **Homogeneidade:** Quando os dados são relativamente homogêneos e não apresentam variações extremas.

**Quando não usar a média aritmética?**

- **Distribuição Assimétrica:** Se os dados são altamente assimétricos (possuem uma cauda longa à esquerda ou à direita), a média pode ser distorcida.
- **Outliers:** Se existem outliers extremos, a média aritmética pode não representar bem a tendência central dos dados.
- **Dados Categóricos:** Quando os dados são categóricos ou qualitativos, a média aritmética não é aplicável.

**A amostra de exemplo é propícia para se utilizar a média aritmética?**

Sim, pois os valores das durações das chamadas VoIP são dados quantitativos e contínuos. Além disso, os valores parecem estar distribuídos de forma relativamente homogênea e não há evidência de outliers extremos que possam distorcer significativamente a média.

**Conclusão**

A análise das durações das chamadas VoIP com base na média aritmética de 2,8 minutos sugere que, em média, as chamadas nesta amostra duraram aproximadamente 2,8 minutos. A média aritmética é uma medida adequada para esta amostra, pois os dados são quantitativos e relativamente homogêneos. No entanto, é importante lembrar que a média aritmética pode ser menos representativa em conjuntos de dados com distribuições altamente assimétricas ou com a presença de outliers extremos.

- **Dados Qualitativos:** Quando os dados são qualitativos e não podem ser ponderados de forma significativa.

#### Média Geométrica

Abaxo está um teste da função `geometric_mean`. A função recebe uma lista de valores 'data' e retorna a média geométrica desses valores.

```python
# Observou-se os seguintes valores de uma variável em um experimento:
data = [18, 13, 11, 8, 10, 58, 12]

mean = geometric_mean(data)

print(f'A média geométrica dos valores observados é: {mean}')
plot_values_with_mean(data, mean)
```

**O que a média geométrica ajuda a identificar/revela?**

- **Crescimento Proporcional:** É uma medida adequada para calcular a taxa de crescimento média de variáveis que crescem de forma proporcional.
- **Mediana Multiplicativa:** Proporciona uma visão mais realista do valor central em conjuntos de dados com grande variabilidade ou outliers, especialmente quando os dados são distribuídos de forma multiplicativa.
- **Redução de Impacto de Outliers:** Ao contrário da média aritmética, a média geométrica reduz o impacto de outliers grandes, pois considera o produto dos valores em vez da soma.

**A média geométrica é indicada para quais tipos de amostra?**

- **Crescimento Exponencial ou Multiplicativo:** Os dados refletem crescimentos ou decrescimentos exponenciais ou multiplicativos.
- **Proporções e Taxas:** É comum em cálculos de médias de proporções, taxas de retorno, ou índices financeiros.

**Quando não usar a média geométrica?**

- **Valores Negativos ou Zeros:** A média geométrica não pode ser calculada com valores negativos ou zeros, pois envolve a raiz n-ésima do produto dos valores.
- **Dados Aditivos:** Quando os dados são somados ou combinados de forma aditiva, a média aritmética é mais apropriada.
- **Distribuição Simétrica Sem Crescimento Multiplicativo:** Para dados que são simétricos e não envolvem crescimento multiplicativo, a média aritmética pode ser mais adequada.

**Esta amostra é propícia para se utilizar a média geométrica?**

Sim, esta amostra pode ser propícia para a utilização da média geométrica, especialmente se os valores observados representam taxas de crescimento ou dados que variam de forma multiplicativa. A média geométrica fornece uma visão mais equilibrada do valor central, especialmente quando há uma alta variabilidade entre os valores, como pode ser visto com a presença do valor 58.

**Conclusão**

A análise dos valores observados no experimento, usando a média geométrica, resultou em um valor de 14,63. Este valor indica o valor central multiplicativo dos dados e representa uma taxa de crescimento média ao longo do experimento. A média geométrica é particularmente útil neste caso, pois proporciona uma visão mais equilibrada do valor central, especialmente devido a presença de um outlier, valor 58.



#### Média Harmônica

Abaixo está um teste da função `harmonic_mean`. A função recebe uma lista de valores 'data' e retorna a média harmônica desses valores.

```python
# Considere o conjunto de medidas (em MFLOPS) obtidas em um experimento. 

data = [405, 367, 405, 419, 388]

mean = harmonic_mean(data)

print(f'O valor médio de MFLOPS é: {mean}')

plot_values_with_mean(data, mean)
```

**O que a média harmônica ajuda a identificar/revela?**

- **Média Ponderada Inversa:** É uma medida de tendência central que leva em consideração a média ponderada inversa dos valores.
- **Média de Taxas ou Razões:** É ideal para calcular médias de taxas, como velocidades ou fluxos, onde as quantidades inversas são somadas.
- **Impacto de Valores Baixos:** A média harmônica é mais sensível a valores baixos, o que a torna útil para evitar a subestimação de médias em conjuntos de dados com valores extremamente baixos.

**A média harmônica é indicada para quais tipos de amostra?**

- Representam taxas, razões ou velocidades (como MFLOPS, km/h, tarefas por hora).
- Possuem valores estritamente positivos (não podem ser zero ou negativos).
- Apresentam grande variabilidade, com valores extremos que podem distorcer outras médias.

**Quando não usar a média harmônica?**

- A amostra contém valores zero ou negativos, pois a fórmula envolve a divisão por cada valor.
- O objetivo é calcular a média de valores que não representam taxas ou razões (como altura, peso, temperatura). Nesses casos, a média aritmética ou a mediana podem ser mais apropriadas.

**Esta amostra é propícia para se utilizar a média harmônica?**

Sim, a amostra de MFLOPS é propícia para o uso da média harmônica, pois:

- Representa uma taxa de execução (operações por segundo).
- Não contém valores zero ou negativos.
- Pode apresentar variabilidade, com algumas medidas mais baixas impactando o desempenho geral.

**Conclusão**

A média harmônica de 395.98 MFLOPS indica a taxa média de execução do sistema, considerando a influência de cada medida individual. Essa medida é mais precisa e representativa do que a média aritmética para este tipo de amostra, pois leva em conta o impacto de valores mais baixos no desempenho geral. A utilização da média harmônica permite uma análise mais completa e acurada do desempenho do sistema em relação à taxa de execução de operações de ponto flutuante.

#### Desvio Padrão

Abaixo está um teste da função `standard_deviation` em um exercício proposto nas Notas de Aula 02. A função recebe uma lista de valores 'data' e retorna o desvio padrão desses valores.

**Exercício (Slide 34 - Aula 2)**

O analista de redes de uma empresa coletou uma amostra do tempo (em segundos) de acesso a uma página web que alguns usuários observaram em dois dias consecutivos. As observações do primeiro dia foram {45, 74, 56, 98, 32, 97} e as do segundo dia foram {48, 81, 33, 40, 84}. Ache a variância amostral e o desvio-padrão de cada amostra. Faça uma análise comparativa.

```python
#Primeiro dia
data = [45, 74, 56, 98, 32, 97]

variance_value = variance(data)
std_deviation_value = standard_deviation(data)

print(f'# Primeiro dia\n')
print(f'Variância amostral do primeiro dia: {variance_value}s²')
print(f'Desvio-padrão do primeiro dia: {std_deviation_value}s')
```

```python
#Segundo dia
data = [48, 81, 33, 40, 84]

variance_value = variance(data)
std_deviation_value = standard_deviation(data)

print(f'# Segundo dia\n')
print(f'Variância amostral do segundo dia: {variance_value}s²')
print(f'Desvio-padrão do segundo dia: {std_deviation_value}s')
```

A variância amostral e o desvio padrão amostral são medidas de dispersão que indicam o quão espalhados estão os dados em torno da média. A variância amostral é calculada como a soma dos quadrados dos desvios em relação à média, dividida pelo número de observações menos um. O desvio padrão amostral é a raiz quadrada da variância amostral.
