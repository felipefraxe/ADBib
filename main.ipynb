## Medidas de Tendência Central

#### Média Aritmética

Abaixo está um teste da função `arithmetic_mean`. A função recebe uma lista de valores 'data' e retorna a média aritmética desses valores.

```python
# Em uma amostra foram registradas a duração (em minutos) de 10 chamadas VoIP: 
data = [2.5, 3.5, 2.8, 1.9, 2.4, 2.6, 3.4, 2.8, 3.6, 2.5]
    
mean = arithmetic_mean(data)

print(f'Duração média de uma chamada VoIP: {mean} min')
plot_values_with_mean(data, mean)
```

#### Média Ponderada

Abaixo está um teste da função `weighted_mean`. A função recebe uma lista de tuplas 'data' onde cada tupla é composta por um valor e seu respectivo peso. A função retorna a média ponderada desses valores.

```python
# A amostra a seguir descreve (notas, créditos) de um estudante.
data = [(7, 4), (8, 2), (6, 2), (9, 2), (8, 4), (9, 4)]

mean = weighted_mean(data)

print(f'O coeficiente de rendimento do estudante é: {mean}')
plot_values_with_mean(data, mean)
```

#### Média Geométrica

Abaxo está um teste da função `geometric_mean`. A função recebe uma lista de valores 'data' e retorna a média geométrica desses valores.

```python
# Observou-se os seguintes valores deuma variável em um experimento:{18, 13, 11, 8, 10, 58, 12}
data = [18, 13, 11, 8, 10, 58, 12]

mean = geometric_mean(data)

print(f'A média geométrica dos valores observados é: {mean}')
plot_values_with_mean(data, mean)
```
